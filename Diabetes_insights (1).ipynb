### Visualizing the value count of Target variable Diabetes_bianry

### Relationship Between HighBP and Diabetes_binary

```python
HighBP_status_counts = df['HighBP'].value_counts()
plt.pie(HighBP_status_counts, labels=HighBP_status_counts.index, autopct='%1.1f%%')
```

```python
# checking the relation between HighBP and Diabetes using barchart
cross_tab = pd.crosstab(df['HighBP'], df['Diabetes_binary_str'])

# Plot the cross-tabulation as a bar chart
cross_tab.plot.bar(figsize=(5, 4))

# Set the title and labels for the plot
plt.title('Diabetes Rate for Those with High Blood Pressure')
plt.xlabel('HighBP')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

Observation: 

HighBP plays an important role in diabetes. As HighBP cases increase, diabetes cases also rise.
